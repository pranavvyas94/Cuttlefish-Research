```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import colors
import matplotlib.pylab as pl
import matplotlib.patches as patches

import os
import time
import shutil
import imageio

from skimage import io, color, img_as_ubyte
```

```python
# path0 = 'E:\Quarter 4.1\Prakash Lab\Squid Paper\High Speed Camera\Beat Frequency\\6\cropped 6\Large set'
path = 'E:\Quarter 4.1\Prakash Lab\Squid Paper\High Speed Camera\Beat Frequency\\39 section good data taken\cropped'

# image_collec0 = os.listdir(path0)
# mean_px_val0 = []
# #for loop to be run over all the images
# for image in image_collec0:
# # for image in image_collec:    
#     print(image)
#     folder_name = [path0,'\\',image]
#     file_name = ''.join(map(str, folder_name))
#     read_image = io.imread(file_name)
    
#     trial = color.rgb2gray(read_image)  #io.imread is better than plt.imread as it prevents data loss in coversion
#     trial8 = img_as_ubyte(trial)

#     mean_px_val0.append(np.min(trial8))
    
image_collec = os.listdir(path)
mean_px_val = []
#for loop to be run over all the images
for image in image_collec:
# for image in image_collec:    
    print(image)
    folder_name = [path,'\\',image]
    file_name = ''.join(map(str, folder_name))
    read_image = io.imread(file_name)
    
    trial = color.rgb2gray(read_image)  #io.imread is better than plt.imread as it prevents data loss in coversion
    trial8 = img_as_ubyte(trial)

    mean_px_val.append(np.min(trial8))
```

```python
path = "E:\Quarter 4.1\Prakash Lab\Squid Paper\Figures" 

# Number of samplepoints
N = 500
# sample spacing
T = 1.0 / 860.0
x = np.linspace(0.0, N*T, N)
y = np.array(mean_px_val)
y0 = np.array(mean_px_val0)
yf = np.fft.fft(y)
yf0 = np.fft.fft(y0)
xf = np.linspace(0.0, 1.0/(2.0*T), int(N/2))  
#according to Nyquist-Shannon Sampling theorem, we can only analyze frequency components up to half of the sampling rate.
#so the maximum frequency of sampling should be 860/2 = 430

fig1, ax1 = plt.subplots(figsize=(15, 8))
ax1.set_title('FFT(Minimum Pixel Value) vs Frequency(Hz)', fontsize = 25)

color_ax = 'steelblue'
ax1.set_xlabel('Frequency (Hz)', fontsize = 25)
ax1.set_ylabel('FFT(Minimum Pixel Value)', fontsize = 25)

# ax1.plot(xf[1:], 2.0/N * np.abs(yf0[:N//2][1:]),color="tab:orange", label = "Series 1")
ax1.plot(xf[1:], 2.0/N * np.abs(yf[:N//2][1:]), color="tab:blue", linewidth = "3", label = "Series 2")

# for i in np.where(np.abs(yf0[1:250]) == np.max(np.abs(yf0[1:250]))):
#     print(i)
#     x_val, y_val = xf[i+1], yf0[i+1]
#     ax1.plot([x_val,x_val],[0,2.0/N * np.abs(y_val)], linewidth = "4", linestyle="dashed",color="tab:orange")    
#     ax1.scatter(x_val,2.0/N * np.abs(y_val), color="tab:orange", s= 75)

for i in np.where(np.abs(yf[1:250]) == np.max(np.abs(yf[1:250]))):
    print(i)
    x_val, y_val = xf[i+1], yf[i+1]
    ax1.plot([x_val,x_val],[0,2.0/N * np.abs(y_val)], linewidth = "5", linestyle="dashed",color="k")
    ax1.scatter(x_val,2.0/N * np.abs(y_val), color="k", s=75)
    
ax1.text(0.22, 0.85,'$41.45 Hz$', horizontalalignment='center',
     verticalalignment='center',
     transform=ax1.transAxes, fontsize=25, color = 'k')

# ax1.text(0.3, 0.95,'$90 Hz$', horizontalalignment='center',
#      verticalalignment='center',
#      transform=ax1.transAxes, fontsize=25, color = 'tab:orange')

ax1.minorticks_on()
ax1.grid(which='major', axis='both')
ax1.grid(which='minor', axis='x')
ax1.tick_params(axis='both', which='major', labelsize=15)
# ax.tick_params(axis='both', which='minor', labelsize=8)

# ax1.legend(bbox_to_anchor=(0.84, 0.7), loc='upper left', borderaxespad=0., fontsize = "15")

save_file_name = os.path.join(path, 'fft_vs_f'+'.tif')
plt.savefig(save_file_name, format='tif', bbox_inches='tight', pad_inches=0.1, dpi = 300)
```
