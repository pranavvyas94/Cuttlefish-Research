```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import colors
import matplotlib.pylab as pl
import matplotlib.patches as patches

import os
import time
import shutil
import imageio

import control.utils as utils

from skimage import io, color, img_as_ubyte
```

```python
path = "F:\Pranav Cuttlefish\\11.07.20 Zscan testing\zstack_2um_150_20-11-07 22-27-44.174913\\0"

image_collec = os.listdir(path)
focus_measure = []
#for loop to be run over all the images
for image in image_collec:
# for image in image_collec:    
    print(image)
    folder_name = [path,'\\',image]
    file_name = ''.join(map(str, folder_name))
    read_image = io.imread(file_name)
    
    trial = color.rgb2gray(read_image)  #io.imread is better than plt.imread as it prevents data loss in coversion
    trial8 = img_as_ubyte(trial)
    
    focus_measure.append(utils.calculate_focus_measure(trial8))
    
```

```python
path = "E:\Quarter 4.1\Prakash Lab\Squid Paper\Figures" 

fig1, ax1 = plt.subplots(figsize=(15, 8))
ax1.set_title('Focus Measure vs Z_step', fontsize = 25)

color_ax = 'steelblue'
ax1.set_xlabel('Z_step', fontsize = 25)
ax1.set_ylabel('Focus Measure', fontsize = 25)

z_range = np.linspace(0,149,150)
ax1.scatter(z_range, focus_measure, color="tab:blue", label='Series 2', linewidth = "3")


ax1.minorticks_on()
ax1.grid(which='major', axis='both')
# ax1.grid(which='minor', axis='both')
ax1.tick_params(axis='both', which='major', labelsize=15)
# ax.tick_params(axis='both', which='minor', labelsize=8)

# ax1.legend(bbox_to_anchor=(0.84, 0.7), loc='upper left', borderaxespad=0., fontsize = "15")

# save_file_name = os.path.join(path, 'apv_vs_t'+'.tif')
# plt.savefig(save_file_name, format='tif', bbox_inches='tight', pad_inches=0.1, dpi = 300)
```
